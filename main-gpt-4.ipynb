```python
from lldm.campaign import Campaign
from lldm.prompter import Prompter
import json

campaign = Campaign(
    campaign='chultanhos',
    session='20230911',
)
```

```python
recap = Prompter(
    gpt4=True,
    data=campaign.data, 
    kinds=[
        "recap",
        "campaign",
        "session",
    ]
)
chat = recap.chat()
print(chat)
```

```python
current_scene = {
    "location": "Hamta's compound: a hallway leading to the slave pens",
    "type": "encounter",
    "title": "The Minotaur",
    "moment": """
        The party is face to face with a minotaur, the minotaur has seen them before and draws his sword. There will be no talking.
    """, 
}

possible_outcomes = {
    "easy dispatch": "The minotaur is quickly dispatched, and the party continues on their way.",
    "hard dispatch": "The minotaur is quickly dispatched, but not before he sounds the alarm. The party continues on their way, but they will have to be more careful.",
    "capture": "The party is defeated and captured. They are taken to the slave pens."
}
recap = chat
context = """
    They are deep in Hamta's compound, twenty yards from the slave pens, where they believe Bero is being held. They got here
    by navigating a secret passage that apparently Hamta and his guards are not aware of. According to some cats that they met
    (Oggvay can talk to cats), passageways like this exist all over Bodat, and more often then not, the minotaurs are not aware of them.
"""

def prepare_scene_data(
        recap: str,
        context: str,
        scene: dict,
        possible_outcomes: dict
):
    scene["recap"] = recap
    scene["context"] = context
    scene["possible_outcomes"] = possible_outcomes
    return scene

scene = prepare_scene_data(
    recap=recap,
    context=context,
    scene=current_scene,
    possible_outcomes=possible_outcomes
)
    
```

```python
scener = Prompter(
    data=scene, 
    kind="scenes",
)
scenes = json.loads(scener.chat())

```

```python
next_scene = scenes[2]
next_scene = prepare_scene_data(
    recap=recap,
    context=context,
    scene=next_scene,
    possible_outcomes={
        "good": "They enter the room and find many treasures and wondrous items",
        "neutral": "They continue on their way",
        "bad": "They enter the room and trigger a magic trap"
    }
)
```

```python
scener = Prompter(
    data=next_scene, 
    kind="scenes",
)
scenes = json.loads(scener.chat())
scenes
```

```python
for scene in scenes:
    print(scene["previous_outcome"].upper(), scene["type"].upper(), scene["title"])
    print(scene["location"])
    print(scene["moment"])
    print()
    print()
```

```python
# Initialize a 3x3 matrix with empty lists
matrix = [[[] for _ in range(3)] for _ in range(3)]

# Define the mapping for scene types and outcomes to matrix indices
scene_type_mapping = {"interaction": 0, "obstacle": 1, "encounter": 2}
outcome_mapping = {"good": 0, "neutral": 1, "bad": 2}

# Populate the matrix
for scene in scenes:
    scene_type_index = scene_type_mapping[scene["type"]]
    outcome_index = outcome_mapping[scene["previous_outcome"]]
    matrix[scene_type_index][outcome_index].append(scene)


for i in range(3):
    for j in range(3):
        if matrix[i][j]:
            print("{:<30} {:<30}".format(matrix[i][j][0]['title'], matrix[i][j][0]['location']), end="")
        else:
            print("{:<30} {:<30}".format("None", "None"), end="")
    print()
```
